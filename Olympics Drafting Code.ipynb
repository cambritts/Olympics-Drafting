```python
import random

list_1=['United States', 'Germany', 'United Kingdom', 'France', 'Italy']
list_2=['China', 'Sweden', 'Japan', 'Norway', 'Australia']
list_3=['Russia', 'Canada', 'Hungary', 'Finland', 'Netherlands']
list_4=['Switzerland', 'South Korea', 'Austria', 'Poland', 'Romania']

combined_list = list_1 + list_2 + list_3 + list_4

while combined_list:
    random_country_1 = random.choice(combined_list)
    combined_list.remove(random_country_1)
    
    if random_country_1 in list_1:
        if list_4:
            random_country_2 = random.choice(list_4)
            list_4.remove(random_country_2)
            if random_country_2 in combined_list:
                combined_list.remove(random_country_2)
        else:
            random_country_2 = None
    elif random_country_1 in list_4:
        if list_1:
            random_country_2 = random.choice(list_1)
            list_1.remove(random_country_2)
            if random_country_2 in combined_list:
                combined_list.remove(random_country_2)
        else:
            random_country_2 = None
    elif random_country_1 in list_2:
        if list_3:
            random_country_2 = random.choice(list_3)
            list_3.remove(random_country_2)
            if random_country_2 in combined_list:
                combined_list.remove(random_country_2)
        else:
            random_country_2 = None
    else:
        if list_2:
            random_country_2 = random.choice(list_2)
            list_2.remove(random_country_2)
            if random_country_2 in combined_list:
                combined_list.remove(random_country_2)
        else:
            random_country_2 = None

    print("Random Country 1: ", random_country_1)
    print("Random Country 2: ", random_country_2)

    if not combined_list or (random_country_1 in list_1 and not list_4) or (random_country_1 in list_4 and not list_1) or (random_country_1 in list_2 and not list_3) or (random_country_1 in list_3 and not list_2):
        break

    print(" ")
```
